# Práctica de Python: If y Else

En esta práctica aprenderás a usar las estructuras condicionales **if** y **else** en Python.

Recuerda:
```python
if condición:
    # código si la condición es verdadera
else:
    # código si la condición es falsa
```

## Ejercicio 1
Pide al usuario que ingrese su edad y muestra:
- 'Eres mayor de edad' si tiene 18 años o más.
- 'Eres menor de edad' si tiene menos de 18.

Escribe tu código en la celda siguiente:

````python
edad=int(input("ingresa tu edad"))
if edad>18:
  print("eres mayor de edad")
elif edad<18:
  print("eres menor de edad")
else:
  print("tienes 18 años")

````

## Ejercicio 2
Pide al usuario que ingrese un número. Si el número es **positivo** muestra 'Es positivo', si es **negativo** muestra 'Es negativo', si es **cero** muestra 'Es cero'.

Escribe tu código en la celda siguiente:

````python
numero=int(input("pon un numero"))
if numero>0:
   print("numero positivo")
elif numero<0:
   print("numero negativo")
else:
   print(" el numero es cero")
````
